```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
single = pd.read_csv("single.csv", sep=";")
two = pd.read_csv("two.csv", sep=";")
four = pd.read_csv("four.csv", sep=";")
eight = pd.read_csv("eight.csv", sep=";")
twelve = pd.read_csv("twelve.csv", sep=";")
sixteen = pd.read_csv("sixteen.csv", sep=";")
twenty = pd.read_csv("twenty.csv", sep=";")
```

```python
field_sizes = [10,100,1000]
plt.figure(figsize = (15,5))
axes = plt.axes()
axes.set_yscale("linear")
axes.set_xscale("log")
axes.set_xlabel('Number of cells in a row')
axes.set_ylabel('Time(ns)')
plt.grid(True)
plt.plot(field_sizes, single['time'], 'o--',
         field_sizes, two['time'], '+--',
         field_sizes, four['time'], '^--',
         field_sizes, eight['time'], 's--',
         field_sizes, twelve['time'], '>--',
         field_sizes, sixteen['time'], '<--',
         field_sizes, twenty['time'], 'b*--')
plt.legend(['Single','2 Threads','4 Threads','8 Threads','12 Threads','16 Threads','20 Threads'])
plt.title('Thread dependency')
plt.show()
```

```python
plot_1 = plt.plot(field_sizes, single['time'], label='Single')
plot_2, = plt.plot(field_sizes, two['time'], label='2')
plot_4, = plt.plot(field_sizes, four['time'], label='4')
plot_8, = plt.plot(field_sizes, eight['time'], label='8')
plot_12, = plt.plot(field_sizes, twelve['time'], label='12')
plot_16, = plt.plot(field_sizes, sixteen['time'], label='16')
plot_20, = plt.plot(field_sizes, twenty['time'], label='20')
plt.legend([plot_1,plot_2,plot_4,plot_8,plot_12,plot_16,plot_20])
plt.show()
```
